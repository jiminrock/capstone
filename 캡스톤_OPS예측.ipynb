```python
import pandas as pd

# 데이터 시각화를 위한 라이브러리
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

# 필요한 데이터 로드

```python
data_2018 = pd.read_excel('data/2018data.xlsx')
data_2019 = pd.read_excel('data/2019data.xlsx')
data_2020 = pd.read_excel('data/2020data.xlsx')
data_2021 = pd.read_excel('data/2021data.xlsx')
data_2022 = pd.read_excel('data/2022data.xlsx')
```

# 공통된 데이터 추출

```python
# 각 데이터프레임에서 이름을 집합으로 추출

names1 = set(data_2018['이름'])
names2 = set(data_2019['이름'])
names3 = set(data_2020['이름'])
names4 = set(data_2021['이름'])
names5 = set(data_2022['이름'])
```

```python
# 모든 집합의 교집합을 찾음
common_names = names1.intersection(names2, names3, names4, names5)

# 공통 이름을 가진 행만 필터링
filtered_df1 = data_2018[data_2018['이름'].isin(common_names)]
filtered_df2 = data_2019[data_2019['이름'].isin(common_names)]
filtered_df3 = data_2020[data_2020['이름'].isin(common_names)]
filtered_df4 = data_2021[data_2021['이름'].isin(common_names)]
filtered_df5 = data_2022[data_2022['이름'].isin(common_names)]
```

```python
# 필터링된 데이터프레임을 하나로 합침
result_df = pd.concat([filtered_df1, filtered_df2, filtered_df3, filtered_df4, filtered_df5]).reset_index(drop=True)
```

```python
result_sort = result_df.sort_values(by = ['이름', '팀'])
result_sort
```

# 데이터프레임에서 필요없는 열 제거

```python
df = result_sort.drop(['순', 'G', '도루','안타','2타','3타','홈런', '사구', '도실', '희타', '희비', '병살', '삼진', '고4'], axis = 1)
df
```

```python
df.reset_index()
```

```python
# 데이터프레임을 csv 파일로 저장하는 방법
df.to_csv('5개년_ops.csv')
```

# OPS와 이것저것의 양의 상관관계 시각화 도움자료

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows = 3)
fig.set_size_inches(18,12)

plt.xticks(rotation = 30, ha = 'right')
sns.pointplot(data = df, x = df['타율'][:50], y = 'OPS', ax = ax1)
sns.pointplot(data = df, x = df['루타'][:50], y = 'OPS', ax = ax2)
sns.pointplot(data = df, x = df['타점'][:50], y = 'OPS', ax = ax3)
```
